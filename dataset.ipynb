```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
from random import *
```

```python
img_height = 64

img_width = 64

background_color = (255, 255, 255)

min_center_offset_from_frame = 8

min_axis_size = 5

max_axis_size = ((img_height + img_width) / 2) / 2
```

```python
def create_background():
    # Create a black image
    img = np.zeros((img_height, img_width, 3), np.uint8)

    # White background
    background_color = (255, 255, 255)
    img[::] = (background_color[0], background_color[1], background_color[2]) 
    
    return img    
```

```python
def create_ellipse(img,
                   center,
                   axes,                   
                   angle,
                   color):
    startAngle = 0
    endAngle = 360
    
    cv2.ellipse(img, center, axes, angle, 0, 360, color, -1)

    return img
```

```python
def create_random_ellipse(img):    
    center_x = randint(1+min_center_offset_from_frame, img_width-min_center_offset_from_frame)-1
    center_y = randint(1+min_center_offset_from_frame, img_height-min_center_offset_from_frame)-1
    center = (center_x, center_y)

    axis_one = randint(min_axis_size, max_axis_size)-1
    axis_two = randint(min_axis_size, max_axis_size)-1
    
    #let's make major_axis >= minor_axis
    major_axis = max(axis_one, axis_two);
    minor_axis = min(axis_one, axis_two);
    
    axes = (major_axis, minor_axis)

    angle = randint(1, 360)-1

    r = randint(1, 256)-1
    g = randint(1, 256)-1
    b = randint(1, 256)-1
    color = (r, g, b)

    img = create_ellipse(img, center, axes, angle, color)
    return img, center, axes, angle
```

```python
def create_set(prefix, size):
    for i in range(0, size):
        img = create_background();

        #Encode the (x_center, y_center), (major_axis, minor_axis) and angle in the name
        
        #For example, if the image contains ellipse:
        #ellipse_00018_T_43_32_21_10_145.png
        #________00018______________________ - serial image id
        #______________T____________________ - contains ellipse
        #________________43_________________ - x_center
        #___________________32______________ - y_center
        #______________________21___________ - major_axis
        #_________________________10________ - minor_axis
        #____________________________145____ - angle

        #If the image doesn't contain ellipse:
        #ellipse_00031_F.png
        #________00031______ - serial image id
        #______________F____ - doesn't contain ellipse
        
        index = format(i, '05d')        
        if randint(0, 1) == 0:
            img, center, axes, angle = create_random_ellipse(img)
            filename = prefix+'/ellipse_'+index+'_T_'+str(center[0])+'_'+str(center[1])+'_'+str(axes[0])+'_'+str(axes[1])+'_'+str(angle)+'.png'    
        else:
            filename = prefix+'/ellipse_'+index+'_F.png'

        print('Creating: '+filename)

        cv2.imwrite(filename, img)
```

```python
train_set_size = 50000
test_set_size = 10000
validation_set_size = 10000

create_set('train', train_set_size)
create_set('test', test_set_size)
create_set('validation', validation_set_size)
```
